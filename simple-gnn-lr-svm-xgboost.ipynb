```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, global_mean_pool, global_max_pool
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score
from torch_geometric.data import Data
from torch_geometric.loader import DataLoader
import pandas as pd
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def create_geometric_data(df):
    num_nodes = len(df)
    x_all = torch.tensor(df.iloc[:, :-1].values, dtype=torch.float)  # All features except 'phishing'
    y_all = torch.tensor(df['phishing'].values, dtype=torch.long)    # 'phishing' labels
    
    data_list = []
    
    # Create a graph for each node
    for i in range(num_nodes):
        # Get the feature and label for the current node
        x = x_all[i:i+1]  # The feature for the current node
        y = y_all[i:i+1]  # The label for the current node
        
        # For a single node, there are no edges
        edge_index = torch.empty((2, 0), dtype=torch.long)  # No edges in a single node graph
        
        # Create the Data object for this node
        data = Data(x=x, edge_index=edge_index, y=y)
        data_list.append(data)
    
    return data_list

train_geometric_data = create_geometric_data(train_data)
test_geometric_data = create_geometric_data(test_data)
```

```python
train_loader = DataLoader(train_geometric_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_geometric_data, batch_size=32)
```

```python
class GCNEmbedding(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=64):
        super(GCNEmbedding, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = self.conv2(x, edge_index)
        x = global_max_pool(x, data.batch)  # Apply mean pooling
        return x
```

```python
gcn_model = GCNEmbedding(input_dim=train_data.shape[1] - 1).to(device)
optimizer = torch.optim.Adam(gcn_model.parameters(), lr=0.01, weight_decay=5e-4)
```

```python
from tqdm import tqdm

def train_gcn(model, loader, optimizer, device):
    model.train()
    for data in tqdm(loader):
        data = data.to(device)
        optimizer.zero_grad()
        embeddings = model(data)
        loss = F.cross_entropy(embeddings, data.y)
        loss.backward()
        optimizer.step()
    print(f"Loss: {loss.item()}")

epochs = 10
for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")
    train_gcn(gcn_model, train_loader, optimizer, device)
```

```python
def extract_embeddings(model, loader):
    model.eval()
    embeddings, labels = [], []
    for data in loader:
        data = data.to(device)
        emb = model(data).cpu().detach().numpy()
        embeddings.append(emb)
        labels.append(data.y.cpu().detach().numpy())
    return np.vstack(embeddings), np.hstack(labels)

train_embeddings, train_labels = extract_embeddings(gcn_model, train_loader)
test_embeddings, test_labels = extract_embeddings(gcn_model, test_loader)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, log_loss
from scipy.stats import norm
```

```python
def calculate_aic_bic(model, X, y):
    predictions = model.predict_proba(X)[:, 1]
    log_likelihood = -log_loss(y, predictions, normalize=False)
    n = len(y)
    k = len(model.coef_.flatten()) + 1 if hasattr(model, 'coef_') else model.n_features_in_
    
    # Calculate AIC
    aic = 2 * k - 2 * log_likelihood
    
    # Calculate BIC
    bic = k * np.log(n) - 2 * log_likelihood
    return aic, bic
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(probability=True),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss")
}
```

```python
model_metrics = {}

for model_name, model in models.items():
    # Train the model
    model.fit(train_embeddings, train_labels)
    
    # Predict on the test set
    test_predictions = model.predict(test_embeddings)
    test_probabilities = model.predict_proba(test_embeddings)[:, 1]
    
    # Calculate metrics
    accuracy = accuracy_score(test_labels, test_predictions)
    f1 = f1_score(test_labels, test_predictions)
    auc_roc = roc_auc_score(test_labels, test_probabilities)
    aic, bic = calculate_aic_bic(model, test_embeddings, test_labels)
    
    # Save the metrics for this model type
    model_metrics[model_name] = {
        "AIC": aic,
        "BIC": bic,
        "Accuracy": accuracy,
        "F1 Score": f1,
        "AUC ROC": auc_roc
    }
```

```python
for model_name, metrics in model_metrics.items():
    print(f"\n==== Best Metrics for {model_name} ====")
    print('AIC:', metrics["AIC"])
    print('BIC:', metrics["BIC"])
    print('Accuracy:', metrics["Accuracy"])
    print('F1 Score:', metrics["F1 Score"])
    print('AUC ROC:', metrics["AUC ROC"])
    print('====\n')
```

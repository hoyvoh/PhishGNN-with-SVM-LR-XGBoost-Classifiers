```python
import dns.resolver
import socket
import requests
import time
import whois
import ssl
from bs4 import BeautifulSoup

def has_spf_record(domain):
    try:
        # Query the TXT records of the domain
        answers = dns.resolver.resolve(domain, 'TXT')
        for record in answers:
            if 'v=spf1' in str(record):
                return 1  # SPF record exists
        return 0  # No SPF record found
    except dns.resolver.NoAnswer:
        return 0  # No SPF record found
    except dns.resolver.NXDOMAIN:
        return 0  # Domain does not exist
    except Exception as e:
        return 0  # Unable to determine SPF presence
        
# Helper functions to count characters in a section
def count_char(section, char):
    return section.count(char) if section else 0

def get_ip_from_url(url):
    try:
        domain = url.split("//")[-1].split("/")[0]
        ip_address = socket.gethostbyname(domain)
        return ip_address
    except Exception as e:
        print("Error getting IP:", e)
        return -1
import requests

# asn
def get_asn_info(ip_address):
    # Replace 'YOUR_API_KEY' with your actual IPinfo API key
    url = f"https://ipinfo.io/{ip_address}/json?token=7f7b4ee2f437cc"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        asn_info = data.get("org")  # This often contains ASN info
        return asn_info
    else:
        print("Failed to retrieve ASN information")
        return -1
def get_asn_team_cymru(ip_address):
    try:
        query = f"{ip_address}.origin.asn.cymru.com"
        response = socket.gethostbyname(query)
        return response  # It returns an ASN along with some additional details
    except Exception as e:
        print("Error retrieving ASN:", e)
        return -1
        
# features['time_response'] = None  # Requires timing info
def get_time_response(url):
    start_time = time.time()
    try:
        response = requests.get(url, timeout=5)  # Adjust timeout as necessary
        end_time = time.time()
        return end_time - start_time  # Time in seconds
    except requests.RequestException as e:
        print("Error fetching URL:", e)
        return -1

# time_domain_activation
def get_time_domain_activation(domain):
    try:
        domain_info = whois.whois(domain)
        return domain_info.creation_date  # May return a list or single date
    except Exception as e:
        print("Error fetching WHOIS data:", e)
        return -1

# time_domain_expiration
def get_time_domain_expiration(domain):
    try:
        domain_info = whois.whois(domain)
        return domain_info.expiration_date  # May return a list or single date
    except Exception as e:
        print("Error fetching WHOIS data:", e)
        return -1

# qty_ip_resolved
def get_qty_ip_resolved(domain):
    try:
        ip_addresses = socket.gethostbyname_ex(domain)[2]
        return len(ip_addresses)  # Count of resolved IPs
    except socket.error as e:
        print(f"Error resolving IP for {domain}: {e}")
        return -1

# qty_nameservers
def get_qty_nameservers(domain):
    try:
        answers = dns.resolver.resolve(domain, 'NS')
        return len(answers)  # Count of nameservers
    except Exception as e:
        print(f"Error fetching nameservers for {domain}: {e}")
        return -1

# qty_mx_servers
def get_qty_mx_servers(domain):
    try:
        answers = dns.resolver.resolve(domain, 'MX')
        return len(answers)  # Count of MX records
    except Exception as e:
        print(f"Error fetching MX servers for {domain}: {e}")
        return -1
        
# ttl_hostname
def get_ttl_hostname(domain):
    try:
        answers = dns.resolver.resolve(domain)
        return answers.ttl  # TTL of the domain
    except Exception as e:
        print(f"Error fetching TTL for {domain}: {e}")
        return -1

# tls_ssl_certificate
def get_tls_ssl_certificate(url):
    try:
        hostname = url.split("//")[-1].split("/")[0]  # Extract the hostname
        port = 443  # Default HTTPS port
        context = ssl.create_default_context()
        with socket.create_connection((hostname, port)) as sock:
            with context.wrap_socket(sock, server_hostname=hostname) as ssock:
                cert = ssock.getpeercert()
                return cert  # Returns the SSL certificate
    except Exception as e:
        print(f"Error fetching SSL certificate for {url}: {e}")
        return -1

# qty_redirects
def count_redirects(url):
    try:
        response = requests.get(url, allow_redirects=True)
        return len(response.history)  # Count of redirects
    except Exception as e:
        print(f"Error fetching redirects for {url}: {e}")
        return -1

# url_google_index
def check_url_google_index(url):
    query = f"site:{url}"
    search_url = f"https://www.google.com/search?q={query}"
    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(search_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # Check if the URL is in the search results
        return any(url in link.get('href') for link in soup.find_all('a'))
    return False

#domain_google_index
def check_domain_google_index(domain):
    query = f"site:{domain}"
    search_url = f"https://www.google.com/search?q={query}"
    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(search_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # Check if the domain is in the search results
        return any(domain in link.get('href') for link in soup.find_all('a'))
    return False
```

```python
import re
import socket
from urllib.parse import urlparse

def extract_features(url, status):
    features = {}

    # Parsing URL components
    parsed_url = urlparse(url)
    domain = parsed_url.netloc
    path = parsed_url.path
    params = parsed_url.query

    # URL-level counts
    features['qty_dot_url'] = url.count('.')
    features['qty_hyphen_url'] = url.count('-')
    features['qty_underline_url'] = url.count('_')
    features['qty_slash_url'] = url.count('/')
    features['qty_questionmark_url'] = url.count('?')
    features['qty_equal_url'] = url.count('=')
    features['qty_at_url'] = url.count('@')
    features['qty_and_url'] = url.count('&')
    features['qty_exclamation_url'] = url.count('!')
    features['qty_space_url'] = url.count(' ')
    features['qty_tilde_url'] = url.count('~')
    features['qty_comma_url'] = url.count(',')
    features['qty_plus_url'] = url.count('+')
    features['qty_asterisk_url'] = url.count('*')
    features['qty_hashtag_url'] = url.count('#')
    features['qty_dollar_url'] = url.count('$')
    features['qty_percent_url'] = url.count('%')
    features['qty_tld_url'] = len(urlparse(url).hostname.split('.')[-1]) if '.' in urlparse(url).hostname else 0
    features['length_url'] = len(url)

    # Domain-level counts
    features['qty_dot_domain'] = domain.count('.')
    features['qty_hyphen_domain'] = domain.count('-')
    features['qty_underline_domain'] = domain.count('_')
    features['qty_slash_domain'] = domain.count('/')
    features['qty_questionmark_domain'] = domain.count('?')
    features['qty_equal_domain'] = domain.count('=')
    features['qty_at_domain'] = domain.count('@')
    features['qty_and_domain'] = domain.count('&')
    features['qty_exclamation_domain'] = domain.count('!')
    features['qty_space_domain'] = domain.count(' ')
    features['qty_tilde_domain'] = domain.count('~')
    features['qty_comma_domain'] = domain.count(',')
    features['qty_plus_domain'] = domain.count('+')
    features['qty_asterisk_domain'] = domain.count('*')
    features['qty_hashtag_domain'] = domain.count('#')
    features['qty_dollar_domain'] = domain.count('$')
    features['qty_percent_domain'] = domain.count('%')
    features['qty_vowels_domain'] = sum([domain.lower().count(v) for v in 'aeiou'])
    features['domain_length'] = len(domain)
    features['domain_in_ip'] = int(bool(re.match(r'^\d+\.\d+\.\d+\.\d+$', domain)))
    features['server_client_domain'] = int("server" in domain or "client" in domain)

    # Directory-level counts
    features['qty_dot_directory'] = path.count('.')
    features['qty_hyphen_directory'] = path.count('-')
    features['qty_underline_directory'] = path.count('_')
    features['qty_slash_directory'] = path.count('/')
    features['qty_questionmark_directory'] = path.count('?')
    features['qty_equal_directory'] = path.count('=')
    features['qty_at_directory'] = path.count('@')
    features['qty_and_directory'] = path.count('&')
    features['qty_exclamation_directory'] = path.count('!')
    features['qty_space_directory'] = path.count(' ')
    features['qty_tilde_directory'] = path.count('~')
    features['qty_comma_directory'] = path.count(',')
    features['qty_plus_directory'] = path.count('+')
    features['qty_asterisk_directory'] = path.count('*')
    features['qty_hashtag_directory'] = path.count('#')
    features['qty_dollar_directory'] = path.count('$')
    features['qty_percent_directory'] = path.count('%')
    features['directory_length'] = len(path)

    # File-level counts
    file_name = path.split('/')[-1]
    features['qty_dot_file'] = file_name.count('.')
    features['qty_hyphen_file'] = file_name.count('-')
    features['qty_underline_file'] = file_name.count('_')
    features['qty_slash_file'] = file_name.count('/')
    features['qty_questionmark_file'] = file_name.count('?')
    features['qty_equal_file'] = file_name.count('=')
    features['qty_at_file'] = file_name.count('@')
    features['qty_and_file'] = file_name.count('&')
    features['qty_exclamation_file'] = file_name.count('!')
    features['qty_space_file'] = file_name.count(' ')
    features['qty_tilde_file'] = file_name.count('~')
    features['qty_comma_file'] = file_name.count(',')
    features['qty_plus_file'] = file_name.count('+')
    features['qty_asterisk_file'] = file_name.count('*')
    features['qty_hashtag_file'] = file_name.count('#')
    features['qty_dollar_file'] = file_name.count('$')
    features['qty_percent_file'] = file_name.count('%')
    features['file_length'] = len(file_name)

    # Parameters-level counts
    features['qty_dot_params'] = params.count('.')
    features['qty_hyphen_params'] = params.count('-')
    features['qty_underline_params'] = params.count('_')
    features['qty_slash_params'] = params.count('/')
    features['qty_questionmark_params'] = params.count('?')
    features['qty_equal_params'] = params.count('=')
    features['qty_at_params'] = params.count('@')
    features['qty_and_params'] = params.count('&')
    features['qty_exclamation_params'] = params.count('!')
    features['qty_space_params'] = params.count(' ')
    features['qty_tilde_params'] = params.count('~')
    features['qty_comma_params'] = params.count(',')
    features['qty_plus_params'] = params.count('+')
    features['qty_asterisk_params'] = params.count('*')
    features['qty_hashtag_params'] = params.count('#')
    features['qty_dollar_params'] = params.count('$')
    features['qty_percent_params'] = params.count('%')
    features['params_length'] = len(params)
    features['tld_present_params'] = int(any(part.endswith('.' + urlparse(url).hostname.split('.')[-1]) for part in params.split('&')))
    features['qty_params'] = len(params.split('&')) if params else 0

    # Other features
    
    features['phishing'] = 1 if status == 'phishing' else 0
    return features

# Test the function
url = "https://github.com/GregaVrbancic/Phishing-Dataset/blob/master/README.md"
features = extract_features(url, 'good')
print(len(list(features.keys())))
```
